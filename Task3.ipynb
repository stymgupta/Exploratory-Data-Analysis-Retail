**Rectify weak areas to generate profit out of them**

```python
import plotly.express as px
import plotly.graph_objects as go
fig = px.scatter(retail, x='Profit', y='Sub-Category', color='Discount', color_discrete_sequence=px.colors.qualitative.Dark24)
fig.show()
```

By the help of graph we can clearly see that the major reason of loss is because of excess discount given i.e. "Profit in Negative". But in some cases the sales of product may increases due to excess discount but the product is still sold in loss.(Binders is perfect example for this).

Also, we can clearly seen that if the discount is less than 0.3%, the product sell is good amount with a substanical profit and thus, more than 0.3% of discount must be avoided in majority of cases.

```python
fig = px.scatter(retail, x='State', y='Profit', color='Category', color_discrete_sequence=px.colors.qualitative.Dark24, width=1100)
fig.show()
```

```python
fig = px.histogram(retail, x='State',color='Discount', color_discrete_sequence=px.colors.qualitative.Dark24)
fig.show()
```

```python
fig = px.histogram(retail, x='State',color='Category', color_discrete_sequence=px.colors.qualitative.Dark24)
fig.show()
```

**Histogram shows region wise distribution of state and purchase graph**
